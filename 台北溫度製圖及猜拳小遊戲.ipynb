```python
# 下載台北思源黑體並命名taipei_sans_tc_beta.ttf，移至指定路徑
!wget -O TaipeiSansTCBeta-Regular.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download

matplotlib.font_manager.fontManager.addfont('TaipeiSansTCBeta-Regular.ttf') # 將下載的思源黑體字形加入字型管理器中
matplotlib.rc('font', family='Taipei Sans TC Beta') # 設置matplotlib使用台北思源黑體字型
# 以上動作是為了讓Colab進行matplotlib繪圖時可顯示繁體中文，否則會出現亂碼

import matplotlib # 導入matplotlib庫

import matplotlib.pyplot as plt # 導入matplotlib.pyplot以進行後續繪圖工作

# 台北市高低溫數據
months = ['一月', '二月', '三月', '四月', '五月', '六月', '七月', '八月', '九月', '十月', '十一月', '十二月']
highs = [28.6, 31.3, 32.9, 35.9, 35.1, 36.9, 37.4, 37.8, 36.1, 34.8, 31.1, 29.5]
lows = [11.9, 13.7, 12.1, 15.2, 17.9, 22.3, 24.8, 25.1, 19.8, 19.9, 16, 11.6]

plt.figure(figsize=(10, 6)) # 設置圖表的大小(可進行調整)

plt.plot(months, highs, marker='o', label='最高溫', color='red') # 繪製最高溫數據
for i, txt in enumerate(highs):
    plt.annotate(txt, (months[i], highs[i]), textcoords="offset points", xytext=(0,10), ha='center') # 在每個數據點上標註數據值，並顯示在數線上方(10為Y軸的偏移量，可調整)
plt.plot(months, lows, marker='o', label='最低溫', color='blue') # 繪製最低溫數據
for i, txt in enumerate(lows):
    plt.annotate(txt, (months[i], lows[i]), textcoords="offset points", xytext=(0,12), ha='center') # 在每個數據點上標註數據值，並顯示在數線上方(12為Y軸的偏移量，可調整)
plt.title('2019年臺北市每月天氣最高溫、最低溫') # 設置圖表標題
plt.xlabel('月份') # 設置X軸標籤
plt.ylabel('溫度 (°C)',rotation=0, labelpad=25) # 設置Y軸標籤，並使其旋轉90度，再設置labelpad(標籤與軸線的距離)，令其不會與Y軸線重疊，以利閱讀
plt.ylim(10,40) #設定Y軸上下限
plt.legend() # 設置圖例
plt.grid(True) # 啟用圖表格線方便閱讀

plt.show() # 顯示圖表
```

```python
import random # 導入random模組，用於生成隨機數字

# 隨機生成一個範圍在1到100之間的數字
target_number = random.randint(1, 100)

print("猜數字遊戲！範圍是1到100之間的數字。")

while True: #無限迴圈，在Break前不斷循環
    try:
        # 使用者輸入數字
        user_input = int(input("請輸入你猜的數字(1-100之間): "))

        if user_input < 1 or user_input > 100:
            print("請輸入1到100之間的數字！")  # 判斷輸入的數字是否在範圍內
        else:

            if user_input < target_number:
                print("數字太小，再大一點")
            elif user_input > target_number:
                print("數字太大，再小一點") # 判斷輸入的數字與目標數字相比大小
            else:
                print("恭喜你，猜對了！")
                break #若猜中的話就跳出無限迴圈，顯示猜對了並停止程序
    except ValueError: #對於使用者輸入非數字以外的字元新增的偵錯程序
        print("請輸入有效的數字")
```
